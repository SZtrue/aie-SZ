```python
import numpy as np # импорт библиотеки numpy
b = np.array([[1, 2], [4, 5]]) # создаём двумерный массив
b

b.shape # форма массива

с = np.ones((2, 2)) # создание массива из единиц
с

a=np.arange(20, 30, 5)
a

a.dtype # определение типа данных

d=np.arange(30, 40, 5)
d

e=a+d # сложение массивов поэлементно
e

e.min() # минимальный элемент массива

e.max() # максимальный элемент массива

f = np.arange(10) # создание массива из 10 элементов
f

f[2:7] # срез с третьего по седьмой элемента массива
```
